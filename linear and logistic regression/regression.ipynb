# Linear Regression

```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import dataset into dataframe
emp_profile = pd.read_csv("./data/employee_profile.csv")
print(emp_profile.head())
print(emp_profile.shape)
```

```python
# Drawing the scattered plot
plt.scatter(emp_profile["Monthly_Income"], emp_profile["Monthly_Expenses"])
plt.title("Income vs Expenses plot")
plt.xlabel("Monthly Income")
plt.ylabel("Monthly Expenses")
plt.show()
```

```python
plt.scatter(emp_profile["Monthly_Income"], emp_profile["Time_Spent_Reading_Books"])
plt.title("Income vs Time Reading Books Spent Plot")
plt.xlabel("Monthly Income")
plt.ylabel("Time Spent Reading Books")
plt.show()
```
